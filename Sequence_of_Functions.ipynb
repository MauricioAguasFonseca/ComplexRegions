Complex Regions.jl is a very usefull package for woking with Complex Regions. 


In this document we will use them for ploting the sequence of functions:
$C_k= z \overline{z}-(3-ik)z-(3+ik)\overline{z}+8+k^2=0$ and their image under the function $f(z)=\frac{1}{\overline{z}}$. 

$f$ correspond to the inverse with respect the unitary circle.


```julia
using ComplexPlots
using ComplexRegions
using ComplexValues
using Plots
```

```julia
plot(Circle(0,1),xlims=(-1,5), ylims=(-5,5), aspect_ratio=:equal, label=false)
for k in -4:4
    C = Circle(3-k*im,1)
    FC = 1/conj(C)
    plot!(C, label=false)
    plot!(FC, label=false)
end
L= Line(4, direction=im)
c = 1/conj(L)
plot!(c, label = false)
plot!(L, label = false) 
L= Line(2, direction=im)
c = 1/conj(L)
plot!(c, label = false)
plot!(L, label = false) 
```

Now we will graph the sequence on functions and their image under $f(z)$, in the Riemann Sphere

```julia
plot(Spherical(Circle(0,1)), label=false)
for k in -4:4
    C = Circle(3-k*im,1)
    FC = 1/conj(C)
    plot!(Spherical(C), label=false)
    plot!(Spherical(FC), label=false)
end
L= Line(4, direction=im)
c = 1/conj(L)
plot!(Spherical(c), label = false)
plot!(Spherical(L), label = false) 
L= Line(2, direction=im)
c = 1/conj(L)
plot!(Spherical(c), label = false)
plot!(Spherical(L), label = false) 
```
